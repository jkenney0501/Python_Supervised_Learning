# Multiple Linear Regression: Bike Demand Predictions

**Data set summary:** \
<a href = 'https://capitalbikeshare.com/'> Capital Bikeshare</a> is the metro DC bikeshare systems. The have more than 5000 bikes across 600+ stations! Its a simple systems of finding a bike at your prefered station and renting it for a specified perios of time. 

**The problem to solve:**\
With such a large system in place, it's important to predict how many bikes will be needed at each station due to demand.

![](uci.png)

**Attribute Information:**

- instant: record index
- dteday : date
- season : season (1:winter, 2:spring, 3:summer, 4:fall)
- yr : year (0: 2011, 1:2012)
- mnth : month ( 1 to 12)
- hr : hour (0 to 23)
- holiday : weather day is holiday or not (extracted from [Web Link])
- weekday : day of the week
- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
+ weathersit :
- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
- atemp: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
- hum: Normalized humidity. The values are divided to 100 (max)
- windspeed: Normalized wind speed. The values are divided to 67 (max)
- casual: count of casual users
- registered: count of registered users
- cnt: count of total rental bikes including both casual and registered

Why are these attributes important? Becasue some of them will impact your decison to rent a bike or not!!!

**Note:** The original dataset did not have weather info and it was later added by the source cited. The weather info is a particularly useful varibale to predict demand! Temprature has been normalized and is presented in Celsius.

## Plot Observations/Explanations

**Continuous:**

- Demand is not normally distributed and may need a log or similar transformation.
- Temp and demand have a strong correlation. Multicol is most lkley present with atemp.
- Humidity and windpseed have little affect on demand but need more statistical analysis.

**Categorical:**
- Demand per season shows that summer is the highest followed by spring and fall with winter being the lowest.
- Demand per month shows that the summmer, spring and fall montsh are the most popular. This most likley has a high correlation with "season".
- Demand per holiday shows a lesser amount on holidays but this is most likley due to frequency. In other words, there are more "regualar" days being considered here.
- Demand per weekday doesnt seem to really matter, the distribution is really uniform.
- Demand per year trends upward from 2011-12 whch would be expected in a regualr economy and a high traffic area such as DC. Howvever, this can be dropped since there are only two years of data.
- Demand per hour is showing that mornings and later afternoons are the peak hours wich makes sense given the works hours of most and leasure time after work hours. This gives an idea of where to put some rental stations and what times!
- Day of week doesnt seem to have any impact on demand.
- Weather does have a large impact, we can see nice weather to decent and cloudy is the most popular with significant drop off after that. This would alos be exepcted to impact demand as the weather does play a role in bike riding sine it is doen outdoors.

**This will allow us to drop some categorical features: weekday, year (this doesnt matter in the regression), workingday, more to be explored later.**
